### Ejercicio 1
Pide la hora (0-23). Si es menor que 12 muestra 'Es mañana', si no 'Es tarde o noche'.


```python
# Escribe tu código aquí
var1 = int(input('ingresa la hora (0-23):'))
if var1<12:
  print('es mañana')
else:
  print('es tarde o noche')
```

### Ejercicio 2
Pide dos números y muestra cuál es mayor o si son iguales.


```python
# Escribe tu código aquí
var1 = float(input('ingresa el primer número: '))
var2 = float(input('ingresa el segundo número: '))

if var1 > var2:
  print('el primer número es mayor')
elif var2 > var1:
  print('el segundo número es mayor')
else:
  print('ambos números son iguales')
```

### Ejercicio 3
Pide un número entero y muestra si es par o impar.


```python
# Escribe tu código aquí
var1 = int(input('ingrsa un número entero: '))

if var1 % 2 == 0:
  print('es un número par')
else:
  print('es un número impar')
```
